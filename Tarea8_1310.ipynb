```python
class NodoDoble:
    def __init__( self , value , siguiente = None , previo = None ):
        self.data = value
        self.next = siguiente
        self.prev = previo
```

```python
class DoubleLinkedList:
    def __init__( self ):
        self.head = None
        self.tail = None
        self.size = 0

    def get_size( self ):
        return self.size

    def is_empty( self ): #true si la lista estÃ¡ vacia
        return self.size == 0

    def append( self , value ):
        if self.is_empty():
            self.head = NodoDoble( value )
            self.tail = self.head
        else:
            self.tail.next = NodoDoble( value, None , self.tail )
            self.tail = self.tail.next
        self.size += 1

    def transversal( self ):
        curr_node = self.head 
        while curr_node != None:
            print(f"| {curr_node.data} | --> " , end="")
            curr_node = curr_node.next
        print("")

    def reverse_transversal( self ):
        curr_node = self.tail 
        while curr_node != None:
            print( f"| {curr_node.data} | --> " , end="" )
            curr_node = curr_node.prev
        print("")

    def find_from_head( self , value ):
        curr_node = self.head
        while curr_node.data != value:
            
            print( f"| {curr_node.data} | --> " , end="" )
            curr_node = curr_node.next
        print(f"| {value} |")

    def find_from_tail( self , value ):
        curr_node = self.tail
        while curr_node.data != value:
            
            print( f"| {curr_node.data} | --> " , end="" )
            curr_node = curr_node.prev
        print(f"| {value} |")
            
    def remove_from_head( self , value ):
        curr_node = self.head
        while curr_node.data != value:
            curr_node = curr_node.next

        if curr_node.data == value:
            curr_node.prev.next = curr_node.next 
            curr_node.next.prev = curr_node.prev
        self.size -= 1

    def remove_from_tail( self , value ):
        curr_node = self.tail 
        while curr_node.data != value:
            curr_node = curr_node.prev

        if curr_node.data == value:
            curr_node.next.prev = curr_node.prev
            curr_node.prev.next = curr_node.next
        self.size -= 1


```

```python
ldl = DoubleLinkedList()
ldl.append(10)
ldl.append(20)
ldl.append(30)
ldl.append(40)
print( "Tranversal:" )
ldl.transversal()
print( "Reverse Tranversal:" )
ldl.reverse_transversal()
print(f"Size: { ldl.get_size() }")
print( f"Is empty? { ldl.is_empty() }" )
ldl2 = DoubleLinkedList()
ldl2.append(10)
ldl2.append(20)
ldl2.append(30)
ldl2.append(40)
ldl2.append(50)
print( "Find(30) from Head :" )
ldl2.find_from_head(30)
print( "Find(20) from Tail :" )
ldl2.find_from_tail(20)
print( "Remove(30) from Head :" )
ldl.remove_from_head(30)
ldl.transversal()
print( "Remove(20) from Tail :" )
ldl2.remove_from_tail(20)
ldl2.transversal()
```
